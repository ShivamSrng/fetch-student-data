```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
students_marks_dataframe = pd.read_csv('data/students_marks.csv', nrows=10)
students_marks_dataframe.head()
```

```python
students_marks_dataframe['Total'] = students_marks_dataframe['Math score'] + students_marks_dataframe['Reading score'] + students_marks_dataframe['Writing score']
students_marks_dataframe.head()
```

```python
students_marks_dataframe['Average Score'] = students_marks_dataframe['Total'] / 3
students_marks_dataframe.head()
```

```python
plt.bar(students_marks_dataframe['Name'], students_marks_dataframe['Math score'])
plt.xlabel('Name')
plt.ylabel('Math Score')
plt.title('Math Scores')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(students_marks_dataframe['Name'], students_marks_dataframe['Reading score'])
plt.xlabel('Name')
plt.ylabel('Reading Score')
plt.title('Reading Scores')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(students_marks_dataframe['Name'], students_marks_dataframe['Writing score'])
plt.xlabel('Name')
plt.ylabel('Writing Score')
plt.title('Writing Scores')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(students_marks_dataframe['Name'], students_marks_dataframe['Average Score'])
plt.xlabel('Name')
plt.ylabel('Average Score')
plt.title('Average Scores')
plt.xticks(rotation=90)
plt.show()
```
